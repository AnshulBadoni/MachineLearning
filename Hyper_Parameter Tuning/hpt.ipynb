```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import svm,datasets
iris = datasets.load_iris()
```

```python
dir(iris)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(iris.data,iris.target,train_size=0.8)
```

```python
model = svm.SVC(C = 30,gamma = 'auto')
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
from sklearn.model_selection import  cross_val_score
```

```python
cross_val_score(svm.SVC(kernel = 'rbf',C = 10,gamma = 'auto'),iris.data,iris.target,cv = 5)
```

```python
cross_val_score(svm.SVC(kernel = 'linear',C = 10,gamma = 'auto'),iris.data,iris.target,cv = 5)
```

```python
import numpy as np
kernals = ['rbf','linear']
c = [10,20,30]
avg_score = {}
for i in kernals:
    for j in c:
        cscore = cross_val_score(svm.SVC(kernel = i,C = j,gamma = 'auto'),iris.data,iris.target,cv = 5)
        avg_score[i + '-' + str(j)] = np.average(cscore)
avg_score
```

```python
from sklearn.model_selection import GridSearchCV

clf = GridSearchCV(svm.SVC(),{
    # 'C':[1,10,20],
    'C': range(1,5),
    'kernel': ['rbf','linear'],
    'gamma': ['auto',10,20]
}, cv = 5,return_train_score=False)

clf.fit(iris.data,iris.target)
clf.cv_results_
```

```python
df = pd.DataFrame(clf.cv_results_)
df
```

```python
df[['param_C','param_gamma','param_kernel','mean_test_score']]
```

```python
clf.best_score_
```

```python
clf.best_params_
```

```python
from sklearn.model_selection import RandomizedSearchCV

rs = RandomizedSearchCV(svm.SVC(),{
    'C':[1,10,20],
    #'C': range(1,5),
    'kernel': ['rbf','linear'],
    'gamma': ['auto',10,20]
}, cv = 5,return_train_score=False,n_iter=2)

rs.fit(iris.data,iris.target)
pd.DataFrame(rs.cv_results_)[['param_C','param_gamma','param_kernel','mean_test_score']]
```
